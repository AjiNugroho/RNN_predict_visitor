```python
data_times = pd.read_csv('visitor_arena.csv')
```

```python
arena_banyak = data_times[['channel_id']]
```

```python
ones = np.ones(data_times.shape[0])
```

```python
arena_banyak.insert(loc=1,column='count',value=ones)
```

```python
arena_banyak=arena_banyak.groupby(['channel_id']).sum()
```

```python
arena_banyak
```

```python
arena_banyak = arena_banyak.sort_values(by=['count'],ascending=False)
```

```python
arena_banyak.head(50)
```

```python
#arena yang memiliki minimal 7 activitas
arena_banyak = arena_banyak.loc[arena_banyak['count']>=7]
```

```python
#list arena yang layak
list_arena = arena_banyak.index.to_list()
```

```python
list_arena
```

```python
data_times=data_times.loc[data_times['channel_id'].isin(list_arena)]
```

```python
data_times
```

```python
data_times.reset_index(drop=True)
```

```python
def trim(df,col,col_value):
    df_o = pd.DataFrame(columns=['visitor','channel_id','created_date']) #blank
    for col_v in col_value:
        df_x = df.loc[df[col]==col_v][:7]
        df_o = df_o.append(df_x,ignore_index=True)
    
    return df_o
        
```

```python
data_times_trim = trim(data_times,'channel_id',list_arena)
```

```python
data_times_trim
```

```python
data_times = data_times.sort_values(by=['created_date'])
```

```python
series_time = data_times_trim['visitor']
```

```python
series_time = series_time.values.reshape(len(list_arena),7)
```

```python
data_times_clean = pd.DataFrame(data=series_time)
```
