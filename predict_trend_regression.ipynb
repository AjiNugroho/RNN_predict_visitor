# Dengan cara analisis trend

```python
import numpy as np
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
data_times = np.genfromtxt('../visitor_data/data_visitor_ori',delimiter=',')
```

```python
#ambil data selama 7 hari 
df_sample = data_times
```

```python
df_sample
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plotX(x):
    plt.plot(np.arange(0,7),x)
```

```python
np.apply_along_axis(plotX,axis=1,arr=df_sample)
```

# menggunakan linear regression, close-form

```python
def regresi(y):
    #training
    X = np.arange(0,7).reshape(-1,1)
    y = y
    lin_reg.fit(X,y)
    #nilai gradient
    return lin_reg.intercept_
    
    
```

```python
gradient = np.apply_along_axis(regresi,axis=1,arr=df_sample)
```
